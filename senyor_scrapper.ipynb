Let's define a method for our google search with 3 parameters

search_term - the keyword that would be use for scraping the website
api_key - my google api key
cse_id - id of our google custom search 

```python
def google_search(search_term, api_key, cse_id, **kwargs):
    service = build("customsearch", "v1", developerKey=api_key)
    res = service.cse().list(q=search_term, cx=cse_id, **kwargs).execute()
    return res
```

Another library we want to utilize. This can pull out raw data from HTML and XML.

```python
url = "https://www.jotform.com/build/203082642769460"
page = requests.get(url)

soup = BeautifulSoup(page.text, "html.parser")
print(soup)
```

Find all column header that has a class named form-matrix-headers

```python
proftags = soup.findAll("th", {"class": "form-matrix-headers" })
proftags
```

Get the text from the parsed elements

```python
for mytag in proftags:
  print(mytag.get_text())
```
